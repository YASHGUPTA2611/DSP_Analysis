## Cycle Elimination Example

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## Generating a Synthetic Signal

```python
time = np.arange(0, 200)
trend_cycle1 = 200*np.sin(2*np.pi*(1/300)*time + np.random.uniform(0.0, 2*np.pi))
trend_cycle2 = 25*np.cos(2*np.pi*(1/100)*time + np.random.uniform(0.0, 2*np.pi))

cycle1 = 12.5*np.sin(2*np.pi*(1/10)*time + np.random.uniform(0.0, 2*np.pi))
cycle2 = 5*np.cos(2*np.pi*(1/5)*time + np.random.uniform(0.0, 2*np.pi))
cycle3 = 3*np.sin(2*np.pi*(1/(3+1/3.0))*time + np.random.uniform(0.0, 2*np.pi))
cycle4 = 20*np.cos(2*np.pi*(1/(2.5))*time + np.random.uniform(0.0, 2*np.pi))
cycle5 = 5*np.sin(2*np.pi*(1/2)*time + np.random.uniform(0.0, 2*np.pi))


cycle = trend_cycle1 + trend_cycle2 + cycle1 + cycle2 + cycle3 + cycle4
```

```python
plt.cla()
plt.plot(time, cycle)
plt.plot(time, pd.Series(cycle).rolling(10).mean(), label='SMA Filter')
plt.plot(time, trend_cycle1+trend_cycle2, label='Cycle Component')
#plt.plot(time, cycle1 + cycle2 + cycle3 + cycle4 + cycle5, label='Trend Component')
plt.legend()
plt.show()
```
